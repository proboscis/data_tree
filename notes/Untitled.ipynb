```coconut
from data_tree.coconut.convert import ImageDef,PILImage,PILImages,Numpy,Torch,VR_0_1,VR_0_255,VR_None,TensorLike
class VisdomHelper:
    def __init__(self,visdom:Visdom):
        self.vis = visdom
    
    def auto_image(self,img,**kwargs):
        candidates = [
            "numpy,float32,CHW,RGB,0_1",
            "numpy,float32,CHW,L,0_1",
            "numpy,float32,BCHW,RGB,0_1",
            "numpy,float32,BCHW,L,0_1"
        ]
        img = img.any_convert(candidates)
        data_type = img.img_def.data_type
        case data_type:
            match Numpy(_,"CHW",_,_):
                res = self.vis.image(img.data,opts=kwargs)
            match Numpy(_,"BCHW",_,_):
                res = self.vis.images(img.data,opts=kwargs)
        return res
```
